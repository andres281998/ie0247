# Introducción: Modulación AM

En esta sección produciremos la señal AM $s(t)=A_c[1+K_a m(t)]\cos(2 \pi f_ct)$ para representar la señal AM en los dominios de tiempo y la frecuencia, y estudiar el efecto de cambiar cualquiera de los parámetros involucrados.

La señal de mensaje será $m(t)=A_m cos(2 \pi f_mt)$ y la portadora $c(t)=A_c cos(2 \pi f_ct)$. Luego se produce la señal AM $s(t)=A_c[1+K_a m(t)]\cos(2 \pi f_ct)$ en el dominio del tiempo.


```python
Am=1 # Amplitud de la señal mensaje
fm=1000 # Frecuencia de la señal mensaje
Ac=1 # Amplitud de la portadora
fc=10000 # Frecuencia de la portadora
Ka=0.5 # Parámetro de sensibilidad de la amplitud.


## -- No es necesario modificar el código siguiente
## ---- Dominio del tiempo
Tstart=0
Tstop=4/fm
Tstep=0.0001/fm

t=np.arange(Tstart,Tstop,Tstep) # Eje tiempo
mt=Am*np.cos(2*math.pi*fm*t) # Señal mensaje
ct=Ac*np.cos(2*math.pi*fc*t) # Señal portadora
st=(1+Ka*mt)*ct # Señal modulada AM
At=np.abs(Ac*(1+Ka*mt))

f = plt.figure(figsize=(20,10))
ax1 = f.add_subplot(311)
ax3 = f.add_subplot(312)
ax5 = f.add_subplot(313)

# Graficar las señales
ax1.plot(t,mt,'b',label='$m(t)$')
ax1.set_xlabel('tiempo (s)')
ax1.set_ylabel('$m(t)$ (v)')
ax1.legend()
ax1.grid()

ax3.plot(t,ct,'k',label='$c(t)$')
ax3.set_xlabel('tiempo (s)')
ax3.set_ylabel('$c(t)$ (v)')
ax3.legend()
ax3.grid()

ax5.plot(t,At,'b--',label='$+|A(t)|$')
ax5.plot(t,-1*At,'k--', label='$-|A(t)|$')
ax5.plot(t,st,'r',label='$s(t)$')
ax5.set_xlabel('tiempo(s)')
ax5.set_ylabel('$s(t)$ (v)')
ax5.legend()
ax5.grid()

## ---- Dominio de la frecuencia
Tstart=0
Tstop=10000/fc
Tstep=0.2/fc

t=np.arange(Tstart,Tstop,Tstep) # Tiempo
N=len(t)
mt=Am*np.cos(2*math.pi*fm*t) # Mensaje
ct=Ac*np.cos(2*math.pi*fc*t) # Portadora
st=(1+Ka*mt)*ct # Señal modulada AM
At=Ac*(1+Ka*mt)
```

# Modulación AM de señales con múltiples componentes armónicos

Se aplicará modulación AM para la señal $X(t)=A_{m1}cos(2 \pi f_{c1}t)+A_{m2}cos(2 \pi f_{c2}t)+A_{m3}cos(2 \pi f_{c3}t)$

Función en Python para generar $m(t)$:

<div style="text-align: justify;">
<p>

### (a) La frecuencia de cada componente de la señal mensaje: fm1, fm2, y fm3:

- Para esta seccion nos situamos en las gráficas de la 22 a la 27.
- fm1 : Cambiar fm1, fm2 y fm3 afecta la frecuencia de la primera, segunda y tercera (respectivamente) componente armónica de la señal mensaje (m1t, m2t y m3t respectivamente). Se puede observar este cambio en la forma de onda de m1t, m2t y m3t en el dominio del tiempo y en la posición de la componente armónica correspondiente en el espectro de frecuencia.

### (b) fc (Frecuencia de la portadora):

- Para esta seccion nos situamos en las gráficas de la 28 a la 29.
- Cambiar fc afecta la frecuencia de la portadora (ct). Esto se verá reflejado en el dominio del tiempo y en el espectro de frecuencia de la señal modulada (st). Un cambio en fc también afectará la separación entre las bandas laterales en el espectro de frecuencia.

### (c) La amplitud de cada componente de la señal mensaje Am1, Am2, y Am3:

- Para esta seccion nos situamos en las gráficas de la 30 a la 35.
- Cambiar las amplitudes Am1, Am2, y Am3 afecta la amplitud de las componentes armónicas individuales de la señal mensaje (m1t, m2t, m3t). Esto se traduce en cambios en la amplitud de la señal modulada (st) en el dominio del tiempo y en el espectro de frecuencia.

### (d) Ac (Amplitud de la portadora):

- Para esta seccion nos situamos en las gráficas 36 y 37.
- Cambiar Ac afectará la amplitud de la portadora (ct). Esto se reflejará en el dominio del tiempo y en el espectro de frecuencia de la señal modulada (st). Un aumento en Ac también puede afectar la separación entre las bandas laterales en el espectro de frecuencia.

Todo lo anterior se comporta igual que la explicacion del incio de como afecta aumentar o disminuir la frecuencia o la amplitud para cada señal, solo que acá es con respecto al ármonico que le correspondé
</p>
</div>

```python
def func_mt(Am1,fm1,Am2,fm2,Am3,fm3,t):
   m1t=Am1*np.cos(2*math.pi*fm1*t) # Componente 1
   m2t=Am2*np.cos(2*math.pi*fm2*t) # Componente 2
   m3t=Am3*np.cos(2*math.pi*fm3*t) # Componente 3
   mt=m1t+m2t+m3t # Señal combinada
   return mt
```

Función en Python para generar la portadora $c(t)$

```python
def func_ct(Ac,fc,t):
   ct=Ac*np.cos(2*math.pi*fc*t) # Señal portadora
   return ct
```

Función en Python para generar la modulación AM $s(t)$.

```python
def Am_mod(mt,ct,Ka):
   st=(1+Ka*mt)*ct # Modulación AM
   return st
```

Producción y graficación de las señales en el tiempo y la frecuencia:

```python
Am1=4 # Amplitud de la componente 1
fm1=1000 # Frecuencia de la componente 1

Am2=4 # Amplitud de la componente 2
fm2=1000 # Frecuencia de la componente 2

Am3=4 # Amplitud de la componente 3
fm3=1000 # Frecuencia de la componente 3

Ac=2 # Amplitud de la portadora
fc=1000 # Frecuencia de la portadora
Ka=1 # Coeficiente de sensibilidad de amplitud

fm1_3=[fm1,fm2,fm3]

## -- No es necesario modificar el código después de esta línea
## Cálculo de la frecuencia fudamental de la señal, para su graficación.
fm=np.gcd.reduce(fm1_3)

## ---- Dominio del tiempo
Tstart=0
Tstop=4/fm
Tstep=0.0001/fm

t=np.arange(Tstart,Tstop,Tstep) # Tiempo
mt=func_mt(Am1,fm1,Am2,fm2,Am3,fm3,t) # Señal mensaje
ct=func_ct(Ac,fc,t) # Portadora
st=Am_mod(mt,ct,Ka) # Modulación AM
At=np.abs(Ac*(1+Ka*mt))

f = plt.figure(figsize=(20,10))
ax1 = f.add_subplot(321)
ax2 = f.add_subplot(322)
ax3 = f.add_subplot(323)
ax4 = f.add_subplot(324)
ax5 = f.add_subplot(325)
ax6 = f.add_subplot(326)

# Graficar las señales
ax1.plot(t,mt,'b',label='$m(t)$')
ax1.set_xlabel('time (sec)')
ax1.set_ylabel('$m(t)$ (v)')
ax1.legend()

ax3.plot(t,ct,'k',label='$c(t)$')
ax3.set_xlabel('time (sec)')
ax3.set_ylabel('$c(t)$ (v)')
ax3.legend()

ax5.plot(t,At,'b--')
ax5.plot(t,-1*At,'k--')
ax5.plot(t,st,'r',label='$s(t)$')
ax5.set_xlabel('time (sec)')
ax5.set_ylabel('$s(t)$ (v)')
ax5.legend()

## ---- Dominio de la frecuencia
Tstart=0; Tstop=100000/fm; Tstep=0.05/fm
t=np.arange(Tstart,Tstop,Tstep) # Tiempo
N=len(t)
mt=func_mt(Am1,fm1,Am2,fm2,Am3,fm3,t) # Mensaje

Mf = np.abs(np.fft.fft(mt))
freqs = np.fft.fftfreq(mt.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Mf=Mf[idx]/N
ax2.plot(freqs, Mf,'b',label='$M(f)$')
ax2.set_xlabel('freq (Hz)')
ax2.set_ylabel('$S(f)$ (v/Hz)')
ax2.legend()

Tstart=0; Tstop=100000/fc; Tstep=0.25/fc
t=np.arange(Tstart,Tstop,Tstep) # Tiempo
N=len(t)
mt=func_mt(Am1,fm1,Am2,fm2,Am3,fm3,t) # Mensaje
ct=func_ct(Ac,fc,t) # Portadora
st=Am_mod(mt,ct,Ka) # Modulación AM
At=np.abs(Ac*(1+Ka*mt))

Cf = np.abs(np.fft.fft(ct))
freqs = np.fft.fftfreq(ct.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Cf=Cf[idx]/N
ax4.plot(freqs, Cf,'k',label='$C(f)$')
ax4.set_xlabel('frecuencia (Hz)')
ax4.set_ylabel('$C(f)$ (v/Hz)')
ax4.legend()

Sf = np.abs(np.fft.fft(st))
freqs = np.fft.fftfreq(st.size, Tstep)
idx = np.argsort(freqs)
freqs=freqs[idx]
Sf=Sf[idx]/N
ax6.plot(freqs, Sf,'r',label='$S(f)$')
ax6.set_xlabel('frecuencia (Hz)')
ax6.set_ylabel('$S(f)$ (v/Hz)')
ax6.legend()
```

# Conclusiones

- Cambiar las frecuencias de las componentes de la señal mensaje (fm1, fm2, fm3) afecta las frecuencias relativas de las componentes armónicas, lo que se refleja en cambios en el espectro de frecuencia.

- El parámetro de sensibilidad de amplitud (Ka) afecta la amplitud de la señal modulada y su espectro de frecuencia. Un aumento en Ka amplifica la modulación de amplitud, haciendo que la señal modulada responda de manera más pronunciada a las variaciones en la señal mensaje.

- Cambiar las amplitudes de las componentes de la señal mensaje (Am1, Am2, Am3) afecta la amplitud de las componentes armónicas individuales y, por ende, la amplitud general de la señal modulada.

- Cambiar la frecuencia de la portadora (fc) afecta la separación entre las bandas laterales en el espectro de frecuencia. 

- La combinación de múltiples componentes armónicos en la señal mensaje introduce complejidad en la señal modulada. Esto se manifiesta en la forma de onda en el dominio del tiempo y en la distribución espectral en el dominio de la frecuencia.
